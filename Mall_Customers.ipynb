```python
from mpl_toolkits.mplot3d import Axes3D

sns.set_style("white")
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(mall_customers.Age, mall_customers["Annual Income"], mall_customers["Spending Score"], c='blue', s=60)
ax.view_init(30, 185)
plt.xlabel("Age", labelpad = 24)
plt.ylabel("Annual Income (k$)", labelpad = 24)
ax.set_zlabel('Spending Score (1-100)', labelpad = 24)
plt.show()
```

```python
from sklearn.cluster import KMeans

wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(mall_customers.iloc[:,1:])
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value", labelpad = 24)
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS", labelpad = 24)
plt.show()
```

```python
km = KMeans(n_clusters=5)
clusters = km.fit_predict(mall_customers.iloc[:,1:])

mall_customers["label"] = clusters

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import animation

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(mall_customers.Age[mall_customers.label == 0], mall_customers["Annual Income"][mall_customers.label == 0], mall_customers["Spending Score"][mall_customers.label == 0], c='blue', s=60)
ax.scatter(mall_customers.Age[mall_customers.label == 1], mall_customers["Annual Income"][mall_customers.label == 1], mall_customers["Spending Score"][mall_customers.label == 1], c='red', s=60)
ax.scatter(mall_customers.Age[mall_customers.label == 2], mall_customers["Annual Income"][mall_customers.label == 2], mall_customers["Spending Score"][mall_customers.label == 2], c='green', s=60)
ax.scatter(mall_customers.Age[mall_customers.label == 3], mall_customers["Annual Income"][mall_customers.label == 3], mall_customers["Spending Score"][mall_customers.label == 3], c='orange', s=60)
ax.scatter(mall_customers.Age[mall_customers.label == 4], mall_customers["Annual Income"][mall_customers.label == 4], mall_customers["Spending Score"][mall_customers.label == 4], c='purple', s=60)
ax.view_init(30, 185)
plt.xlabel("Age", labelpad = 24)
plt.ylabel("Annual Income (k$)", labelpad = 24)
ax.set_zlabel('Spending Score (1-100)', labelpad = 24)

# def rotate(angle):
#     ax.view_init(azim=angle)
#     rot_animation = animation.FuncAnimation(fig, rotate, frames=np.arange(0,362,2),interval=100)
# rotate(330)

    
plt.show()
```
